```python
from matplotlib import pyplot as plt
import numpy as np
import cv2 
from PIL import Image

# The result returned by cv2.imread is in BGR colour space
img_1 = cv2.imread('000000008021.jpg')
print(type(img_1))
# Deep copy
img_2 = img_1.copy()
# Convert from BGR colour space to RGB colour space
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
# BGR image 
plt.subplot2grid((1,2), (0,0))
plt.imshow(img_1)
# RGB image
plt.subplot2grid((1,2), (0,1))
plt.imshow(img_2)
```

```python
import numpy as np
import cv2 
from PIL import Image
from matplotlib import pyplot as plt

# Usage of PIL,and note the returned image is in RGB (different from cv2)
img_1 = Image.open('000000008021.jpg').convert('RGB')
img_2 = np.asarray(img_1)
# Deep copy
img_3 = img_2.copy()
img_3 = cv2.cvtColor(img_3, cv2.COLOR_RGB2BGR)
# img_3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)
img_3

# RGB image 
plt.subplot2grid((1,2), (0,0))
plt.imshow(img_2)
# BGR image
plt.subplot2grid((1,2), (0,1))
plt.imshow(img_3)
```

```python
# Test for doing grid subplot
from matplotlib import pyplot as plt
# colspan tells the number of grids(in column) you want your current subplot to take up.
# rowspan tells the number of grids(in row) you want your current subplot to take up.
plt.subplot2grid((3, 3), (0, 2), colspan = 2)
plt.subplot2grid((3, 3), (0, 1), colspan = 1)
plt.subplot2grid((3, 3), (0, 0), colspan = 1)
plt.subplot2grid((3, 3), (1, 0), rowspan = 2, colspan = 2)
# plt.subplot2grid((3, 3), (0, 1), colspan = 2)
# plt.subplot2grid((3, 3), (2, 1), colspan = 2)
```
