### Random Translation

```python
# Start Random Translation

# Empty dataframe
df = pd.DataFrame() 

# Do random translation three times for each image
number_of_translation_for_each_image = 3
# Final array for storing all bbox coordinates.
bbox_coordinates_array_random_horizontal_shear_final = np.zeros([9*number_of_translation_for_each_image, 4])
file_path_array_pd = []

# Outter loop for number of transforms done for each image.
for i in range(number_of_translation_for_each_image):
    bbox_coordinates_array_random_horizontal_shear_temp = np.zeros([9, 4])
    for j in range(9):

        file_path_temp = file_path_array[j]
        bbox_coordinate_temp = np.array([bbox_coordinates_array[j]])

        # Read the dicom image
        original_dicom = dicom.dcmread(file_path_temp)
        original_data = original_dicom.pixel_array
        dicom_image_size = original_data.shape
        print("Processing number ", j, " images...", i, " iteration...")
        print(dicom_image_size)
        print(dicom_image_size[0])

        # Path of the saved augmentated images.
        saving_path = r'D:\Studying\RetinaNet_Project_test_agumentations\validation_augmentations\Random_horizontal_shear' 
        # This is the sequence number of the dicom image
        saving_folder = '\\'+ file_path_temp.split('\\')[-1][0:-34]
        # If this directory does not exist, create it at first time.
        if not (os.path.exists(saving_path + saving_folder)):
            os.mkdir(saving_path + saving_folder)

        # Use scikit to duplicate the image channel
        three_channel_image = skimage.color.gray2rgb(original_data)
        img_, bboxes_ = RandomShear(0.2)(three_channel_image.copy(), bbox_coordinate_temp.copy())

        img_random_horizontal_shear_temp = img_[:,:,0].copy()
        print(type(img_random_horizontal_shear_temp[0,0]))
        image_name = r'\temp_dicom_{}_random_horizontal_shear_{}.tif'.format(saving_folder[1:], i)
        Image.fromarray(img_random_horizontal_shear_temp).save(saving_path + saving_folder + image_name)
        # Save current temp file path to the data frame array
        file_path_array_pd.append(saving_path + saving_folder + image_name)
        bbox_coordinates_array_random_horizontal_shear_temp[j, :] = [np.int(bboxes_[0,0]), np.int(bboxes_[0,1]), np.int(bboxes_[0,2]), np.int(bboxes_[0,3])]
    bbox_coordinates_array_random_horizontal_shear_final[9*i: 9*i+9 ,:] = bbox_coordinates_array_random_horizontal_shear_temp

print(bbox_coordinates_array_random_horizontal_shear_final)
df = pd.DataFrame(bbox_coordinates_array_random_horizontal_shear_final)
df['file path'] = file_path_array_pd
print(df)
writer = pd.ExcelWriter('Random_horizontal_shear.xlsx')
df.to_excel(writer)
writer.save()
```
