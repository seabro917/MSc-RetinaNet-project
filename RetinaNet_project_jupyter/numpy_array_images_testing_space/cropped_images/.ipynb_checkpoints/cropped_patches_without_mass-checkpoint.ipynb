```python
# ******* True running program *******

# Generate the cropped images and its corresponding bbox coordinates, and write everything in an Excel file.

from matplotlib import pyplot as plt
import numpy as np
import pydicom as dicom
from PIL import Image, ImageMath
import cv2 
import skimage.color
import os
import random
import pandas as pd
```

### Pre-processing

```python
# Load the original images file path
excel_file_path = r'D:\Studying\RetinaNet_Project_csv\new_version\on_my_own_pc\75train_25val_16test\for_image_patches\training_set_no_mass.xlsx'
data_frame = pd.read_excel(excel_file_path)
image_path_dataframe = data_frame['image_path']
print(image_path_dataframe)
# List for storing all image path
image_path_list = np.array(image_path_dataframe).tolist()
```

```python
# Empty dataframe
df = pd.DataFrame()    
    
# Outter loop for going through all 193 original masses(
for j in range(193):
    
    file_path = image_path_list[j]

    # Read the dicom image
    original_dicom = dicom.dcmread(file_path)
    original_data = original_dicom.pixel_array
    dicom_image_size = original_data.shape
    print(dicom_image_size)
    print(dicom_image_size[0])


    # Save the dicom image in .tif format to make sure lossless.
    Image.fromarray(original_data).save('dicom_tiff_temp.tif')
    dicom_tiff = Image.open('dicom_tiff_temp.tif')  
    print(type(dicom_tiff))
    dicom_tiff_image = np.array(dicom_tiff)

    # Path of the saved cropped images.
    saving_path = r'D:\Studying\RetinaNet_project_image_patches\without_mass' 
    # This is the number of the dicom image
    saving_folder = '\\'+ file_path.split('\\')[-1][0:-34]
    
    # If this directory does not exist, create it at first time.
    if not (os.path.exists(saving_path + saving_folder)):
        os.mkdir(saving_path + saving_folder)
        
    number_of_cropped_images = 2

    # Array for store the names of cropped images for the current dicom image
    cropped_image_name_array_temp = []
    
    # Start cropping
    # Inner loop for how many cropped images we want to generate from each dicom image
    for i in range(number_of_cropped_images):

        
        # Crop the image based on whether it is left or right-side mammogram, to make sure that the patch is not totally black.
        # Right-side mammogram
        if(dicom_tiff_image[20,20] == 0):
            # Fixed size patch 1024*1024 pixel^2
            bottom_right_x = random.randint(dicom_image_size[1] - 50, dicom_image_size[1])
            bottom_right_y = random.randint(dicom_image_size[0] - 1000 ,dicom_image_size[0]-500)
            top_left_x = bottom_right_x - 1024
            top_left_y = bottom_right_y - 1024
        
        # Left-side mammogram
        else:
            # Fixed size patch 1024*1024 pixel^2
            top_left_x = random.randint(0, 500)
            top_left_y = random.randint(600, dicom_image_size[0] / 2 )
            bottom_right_x = top_left_x + 1024
            bottom_right_y = top_left_y + 1024

        
        
        crop_coordinate = (top_left_x, top_left_y, bottom_right_x, bottom_right_y)
        crop_image = dicom_tiff.crop(crop_coordinate)

        image_name = r'\temp_dicom_{}_crop_{}.tif'.format(saving_folder[1:], i)
            
        crop_image.save(saving_path + saving_folder + image_name)
        
        # This is for dataframe store purpose
        cropped_image_name_array_temp.append(saving_path + saving_folder + image_name)


    # Remember to close the image at end of each loop
    dicom_tiff.close()
    df_temp = pd.DataFrame(cropped_image_name_array_temp)
    # Append temp dataframe to the total dataframe. 
    df = df.append(df_temp)
print(df)
writer = pd.ExcelWriter('write_xlsx.xlsx')
df.to_excel(writer)
writer.save()
```
