```python
# ******* True running program *******

# Generate the cropped images and its corresponding bbox coordinates, and write everything in an Excel file.

from matplotlib import pyplot as plt
import numpy as np
import pydicom as dicom
from PIL import Image, ImageMath
import cv2 
import skimage.color
import os
import random
import pandas as pd
```

### Pre-processing

```python
# Load the original images file path
excel_file_path = r'D:\Studying\RetinaNet_Project_csv\new_version\on_my_own_pc\75train_25val_16test\for_image_patches\training_set_only_mass_delete_some.xlsx'
data_frame = pd.read_excel(excel_file_path)
image_path_dataframe = data_frame['image_path']
print(image_path_dataframe)
# List for storing all image path
image_path_list = np.array(image_path_dataframe).tolist()
# List for storing their corresponding bounding boxes coordinates
bbox_coordinates_dataframe = pd.read_excel(excel_file_path, usecols=[1,2,3,4], names=None)
print(bbox_coordinates_dataframe)
bbox_coordinates_list = np.array(bbox_coordinates_dataframe).tolist()
```

```python
# Empty dataframe
df = pd.DataFrame()    
    
# Outter loop for going through all 75 original masses(number of image is smaller than 75 since some image contain multiple masses.)
for j in range(74):
    
    file_path = image_path_list[j]
    bbox_coordinate = bbox_coordinates_list[j]

    # Read the dicom image
    original_dicom = dicom.dcmread(file_path)
    original_data = original_dicom.pixel_array
    dicom_image_size = original_data.shape
    print(dicom_image_size)
    print(dicom_image_size[0])
    # Flag to show whther the current image has two or more masses.
    exist_flag = False


    # Save the dicom image in .tif format to make sure lossless.
    Image.fromarray(original_data).save('dicom_tiff_temp.tif')
    dicom_tiff = Image.open('dicom_tiff_temp.tif')  
    print(type(dicom_tiff))
    dicom_tiff_image = np.array(dicom_tiff)

    # Path of the saved cropped images.
    saving_path = r'D:\Studying\RetinaNet_Project_test_agumentations' 
    # This is the number of the dicom image
    saving_folder = '\\'+ file_path.split('\\')[-1][0:-34]
    
    # If this directory does not exist, create it at first time.
    if not (os.path.exists(saving_path + saving_folder)):
        os.mkdir(saving_path + saving_folder)
    else:
        exist_flag = True


    number_of_cropped_images = 5
    # Array for the bbox coordinates of cropped images
    bbox_coordinate_after_cropping_array = np.zeros([number_of_cropped_images, 4])
    
    # Array for store the names of cropped images for the current dicom image
    cropped_image_name_array_temp = []
    
    # Start cropping
    # Inner loop for how many cropped images we want to generate from each dicom image
    for i in range(number_of_cropped_images):

#         # Random cropping and make sure the cropped image contain the bbox.
#         top_left_x = random.randint(0, bbox_coordinate[0])
#         top_left_y = random.randint(0, bbox_coordinate[1])
#         bottom_right_x = random.randint(bbox_coordinate[2], dicom_image_size[1])
#         bottom_right_y = random.randint(bbox_coordinate[3], dicom_image_size[0])
        
          # Fixed size patch 1024*1024 pixel^2
        top_left_x = random.randint(0, bbox_coordinate[0])
        top_left_y = random.randint(0, bbox_coordinate[1])
        bottom_right_x = top_left_x + 1024
        bottom_right_y = top_left_y + 1024
        
        while(bottom_right_x < bbox_coordinate[2] or bottom_right_y < bbox_coordinate[3]):
            top_left_x = random.randint(0, bbox_coordinate[0])
            top_left_y = random.randint(0, bbox_coordinate[1])
            bottom_right_x = top_left_x + 1024
            bottom_right_y = top_left_y + 1024

        crop_coordinate = (top_left_x, top_left_y, bottom_right_x, bottom_right_y)
        crop_image = dicom_tiff.crop(crop_coordinate)
        if(exist_flag == False):
            image_name = r'\temp_dicom_{}_crop_{}.tif'.format(saving_folder[1:], i)
        else:
            image_name = r'\temp_dicom_{}_crop_{}.tif'.format(saving_folder[1:], i+number_of_cropped_images)
        crop_image.save(saving_path + saving_folder + image_name)
        
        # This is for dataframe store purpose
        cropped_image_name_array_temp.append(saving_path + saving_folder + image_name)

        # This part is in cooperation with plotting puepose, can be commented here.
        crop_image = Image.open(saving_path + saving_folder + image_name)
        crop_image_data = np.array(crop_image)


        # Calculate the resulting bbox coordinates after cropping
        bbox_coordinate_after_cropping_temp = [bbox_coordinate[0] - crop_coordinate[0], bbox_coordinate[1] - crop_coordinate[1], bbox_coordinate[2] - crop_coordinate[0], bbox_coordinate[3] - crop_coordinate[1]]

        # Put the current bbox coordinates into array
        bbox_coordinate_after_cropping_array[i, :] = bbox_coordinate_after_cropping_temp

        ####################################################################################################

        ## Plot to see whether the result is correct. Note that this module can be commente here.

        ## Original image and its bbox
        # plt.figure(1)
        # # Plot the dicom image
        # plt.subplot2grid((1,2), (0,0))
        # plt.imshow(dicom_tiff_image, cmap = 'gray')
        # # Plot the bbox in colored filled rectangle.(By setting the end of the command to -1)
        # plt.subplot2grid((1,2), (0,1))
        # cv2.rectangle(dicom_tiff_image, (bbox_coordinate[0], bbox_coordinate[1]), (bbox_coordinate[2], bbox_coordinate[3]), (255, 0, 0), -1)
        # plt.imshow(dicom_tiff_image, cmap = 'gray')
        # # Plot cropped image and its bbox
        # plt.figure(2)
        # plt.subplot2grid((1,2), (0,0))
        # plt.imshow(crop_image_data, cmap = 'gray')
        # # Plot its corresponding bounding box
        # plt.subplot2grid((1,2), (0,1))
        # cv2.rectangle(crop_image_data, (bbox_coordinate_after_cropping_temp[0], bbox_coordinate_after_cropping_temp[1]), (bbox_coordinate_after_cropping_temp[2], bbox_coordinate_after_cropping_temp[3]), (255, 0, 0), -1)
        # plt.imshow(crop_image_data, cmap = 'gray')

        ####################################################################################################
    print(bbox_coordinate_after_cropping_array)
    # Remember to close the image at end of each loop
    dicom_tiff.close()
    df_temp = pd.DataFrame(bbox_coordinate_after_cropping_array)
    # Add a fifth column, which is the file path of the cropped imgaes.
    df_temp['file_name'] = cropped_image_name_array_temp
    # Append temp dataframe to the total dataframe. 
    df = df.append(df_temp)
print(df)
writer = pd.ExcelWriter('write_xlsx.xlsx')
df.to_excel(writer)
writer.save()
```
