```python
# Only mass set
import pandas as pd
import numpy as np

original_csv_path = r'D:\Studying\RetinaNet_Project_csv\new_version\on_my_own_pc\75train_25val_16test\for_patches_training\training_set_only_mass_patches\for_shannon.xlsx'

df = pd.read_excel(original_csv_path)
image_names_list = np.array(df['image_path']).tolist()
print(len(image_names_list))
new_images_path_for_shannon = []

for i in range(375):
    image_path_temp = image_names_list[i]
    reserved_part = image_path_temp.split('s\\')[-1]
    reserved_part = reserved_part.replace('\\', '/')
    new_images_path_temp = '/home/s1966015/RetinaNet_project/INbreast_tif_patches/with_mass' + '/' + reserved_part
    new_images_path_for_shannon.append(new_images_path_temp)

print(new_images_path_for_shannon)

data_frame = pd.DataFrame(new_images_path_for_shannon)
writer = pd.ExcelWriter('shannon.xlsx')
data_frame.to_excel(writer)
writer.save()
```
