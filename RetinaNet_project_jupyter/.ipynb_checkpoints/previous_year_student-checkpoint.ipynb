```python
#encoding=utf-8
import torch
import torchvision
from torchvision import datasets, transforms
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
import torch.optim as optim  
#import matplotlib.pyplot as plt
import numpy as np

from PIL import Image, ImageMath
def my_image_loader(filename):
    with open(filename, 'rb') as f:
        image = Image.open(f)
        image = image.convert(mode='F')
        image = ImageMath.eval("(a-1980.0)/163.0", a=image)
        return image

# Training settings


# MNIST Dataset
#def loadtraindata():
path = "/exports/eddie3_homes_local/dil/bone/train/"
train_dataset = torchvision.datasets.ImageFolder(path,
                                                 loader=my_image_loader,
                                                 transform=transforms.Compose([
                                                 #transforms.Resize(28), 
                                                 transforms.RandomResizedCrop(256),#, padding=4),
                                                 transforms.RandomHorizontalFlip(),
                                                 # transforms.Grayscale(num_output_channels=1),
                                                 transforms.ToTensor()
                                                 #transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1),
                                                 # transforms.Normalize([1980],[163])
                                                 ]
                                                 #transforms.CenterCrop(32)]),
                                                ))


                                                

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=4,
                                              shuffle=True)# , num_workers=2)
    #return train_loader
    
#def loadtestdata():
path = "/exports/eddie3_homes_local/dil/bone/test/"
test_dataset = torchvision.datasets.ImageFolder(path,
                                                 loader=my_image_loader,
                                                 transform=transforms.Compose([
                                                 #transforms.Resize(28), 
                                                 transforms.RandomResizedCrop(256),#, padding=4),
                                                 transforms.RandomHorizontalFlip(),
                                                 # transforms.Grayscale(num_output_channels=1),
                                                 transforms.ToTensor()
                                                 #transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1),
                                                 # transforms.Normalize([1980],[163]),
                                                 ]
                                                 #transforms.CenterCrop(32)]),
                                                ))
                                             
                                                
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=4,
                                             shuffle=True)# , num_workers=0)
    #return test_loader
    
    


class Net(nn.Module):

    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(1, 10, kernel_size=5,stride=1,padding=2)
        self.bn1 = nn.BatchNorm2d(10)
        self.conv2 = nn.Conv2d(10, 20, kernel_size=5,stride=1,padding=2)
        self.bn2 = nn.BatchNorm2d(20)
        self.mp = nn.MaxPool2d(2,2)
        self.fc = nn.Linear(20*64*64, 2)

    def forward(self, x):
        in_size = x.size(0)
        x = F.relu(self.bn1(self.mp(self.conv1(x))))
        x = F.relu(self.bn2(self.mp(self.conv2(x))))
        x = x.view(in_size, -1)  # flatten the tensor
        x = self.fc(x)
        return F.log_softmax(x,dim=1)



model = Net()

optimizer = optim.SGD(model.parameters(), lr=1e-9, momentum=0.9)


def train(epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        data, target = Variable(data), Variable(target)
        # optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        # loss = F.binary_cross_entropy(output, target.float())
        #loss = nn.CrossEntropyLoss()(output, target)
        loss.backward()
        optimizer.step()
        if batch_idx % 4 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.data.item()))


def test():
    model.eval()
    test_loss = 0
    correct = 0
    for data, target in test_loader:
        data, target = Variable(data, volatile=True), Variable(target)
        output = model(data)
        # sum up batch loss
        test_loss += F.nll_loss(output, target, size_average=False).data.item()
        # test_loss += F.binary_cross_entropy(output, target.float())
        # get the index of the max log-probability
        pred = output.data.max(1, keepdim=True)[1]
        correct += pred.eq(target.data.view_as(pred)).cpu().sum()

    test_loss /= len(test_loader.dataset)
    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))

print(model)

for epoch in range(1, 20):
    train(epoch)
    test()
```
