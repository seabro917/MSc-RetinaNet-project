```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pydicom as dicom
import skimage.color
import pandas as pd
```

```python
file_path = r'D:\Studying\RetinaNet_Project_csv\new_version\on_my_own_pc\75train_25val_16test\for_anchor_optimization\training_set_only_mass.xlsx'
data_frame = pd.read_excel(file_path)
image_path_dataframe = data_frame['image_path']
print(image_path_dataframe)
# List for storing all image path
image_path_list = np.array(image_path_dataframe).tolist()
# print(image_path_list[0])
# print(len(image_path_list))
target_folder = r'D:\Studying\RetinaNet_Project_anchor_optimization\75train_25val_16test'
target_image_path_list = []
for i in range(75):
    image_path_temp = image_path_list[i]
    dicom_image = dicom.dcmread(image_path_temp).pixel_array
    print("Processing ", i, "image...")
    image_name_tif = r'temp_dicom_tif_{}.tif'.format(image_path_temp.split('\\')[-1][0:-34])
#     print(target_folder + '\\' + image_name_tif)
    Image.fromarray(dicom_image).save(target_folder + '\\' + image_name_tif)
    target_image_path_list.append(target_folder + '\\' + image_name_tif)


```

```python
print(len(target_image_path_list))
df = pd.DataFrame(target_image_path_list)
print(df)
writer = pd.ExcelWriter('all_mass_training_set.xlsx')
df.to_excel(writer)
writer.save()
```
